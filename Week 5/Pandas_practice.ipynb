## Problem 1

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Score': [85, 92, 78, 88, 95]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
# Select the rows where the 'Age' is greater than 25.
above_25 = df[df['Age'] > 25]
```

```python
# Select the 'Name' and 'Score' columns for the filtered rows.
above_25[['Name', 'Score']]
```

```python
#Calculate the mean 'Score' for the dataframe.

df['Score'].mean()
```

```python
#Group the dataframe by 'City' and calculate the average 'Age' and 'Score' for each city.
df.groupby('City')[['Age','Score']].mean()
```

```python
# Create a new column called 'Pass' that is True if the 'Score' is 80 or above, and False otherwise.
df['Pass'] = df['Score'] >= 80
df
```

```python
# Update the 'Score' column by adding 5 bonus points to each score.
df.loc[:, 'Score'] += 5
df
```

```python
# Sort the dataframe by 'Score' in descending order.
df.sort_values(['Score'], ascending = False)
```

## Problem 2

```python
import pandas as pd
import numpy as np

data = {
     'EmployeeID': range(1, 21),
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack',
             'Kara', 'Leo', 'Mia', 'Nina', 'Owen', 'Paul', 'Quincy', 'Rachel', 'Steve', 'Tina'],
    'Age': [24, 27, 22, 32, 29, 34, 28, 26, 31, 30, 23, 35, 27, 28, 32, 26, 29, 31, 33, 24],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR', 'Finance', 'IT', 'Marketing', 'HR', 'Finance',
                   'IT', 'Marketing', 'HR', 'Finance', 'IT', 'Marketing', 'HR', 'Finance', 'IT', 'Marketing'],
    'Salary': [50000, 60000, 55000, 45000, 48000, 62000, 53000, 47000, 51000, 61000,
               54000, 46000, 49000, 63000, 52000, 48000, 50000, 64000, 57000, 49000],
    'JoinDate': pd.date_range(start='2020-01-01', periods=20, freq='M')
}

df = pd.DataFrame(data)
df
```

```python
# Select employees who are older than 30.

age_over_30 = df[df['Age'] > 30]
age_over_30
```

```python
# Select the 'Name' and 'Salary' columns for these employees.

age_over_30[['Name','Salary']]
```

```python
# Calculate the average 'Salary' for the entire dataframe.

df['Salary'].mean()
```

```python
# Group the dataframe by 'Department' and calculate the average 'Age' and 'Salary' for each department.

df.groupby('Department')[['Age', 'Salary']].mean()
```

```python
# Add a column 'YearsAtCompany' that calculates the number of years 
# each employee has been at the company based on the current date 
# (assume the current date is '2024-08-06').


date_string = '2024-08-06'
date_object = pd.to_datetime(date_string)
df['YearsAtCompany'] = ((date_object - df['JoinDate']).dt.days / 365.25).round(2)
df
```

```python
# Create a new column called 'HighEarner' that is True if the 'Salary' is 60000 or above, and False otherwise.

df['HighEarner'] = df['Salary'] >= 60000
df
```

```python
# Increase the 'Salary' by 10% for all employees in the 'IT' department.

df.loc[df['Department']=='IT', 'Salary'] *= 1.1
df
```

```python
# Sort the dataframe by 'JoinDate' in ascending order.

df.sort_values('JoinDate')
```

```python
# Create another dataframe departments with 'Department' and 'Manager' columns, and merge it with the original dataframe on the 'Department' column.

departments = pd.DataFrame({
    'Department': ['HR', 'Finance', 'IT', 'Marketing'],
    'Manager': ['Laura', 'James', 'Nina', 'Robert']
})

df = pd.merge(df, departments, on='Department', how='left')
df
```
