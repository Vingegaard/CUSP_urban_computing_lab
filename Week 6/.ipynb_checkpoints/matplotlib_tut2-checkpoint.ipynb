## Gas price data

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices Movement (in USD)', fontdict={'fontweight':'bold', 'fontsize': 9})

plt.plot(gas.Year, gas.USA, 'b.-', label = 'US')
plt.plot(gas.Year, gas.Canada, 'r.-', label = 'CAN')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label = 'KOR')
plt.plot(gas.Year, gas.Australia, 'y.-', label = 'AUS')

#countries_of_interest = ['USA', 'Canada', 'South Korea', 'Australia']
#for country in countries_of_interest:
#    plt.plot(gas.Year, gas[country], label=country)
#for country in gas.columns[1:]:
    #plt.plot(gas.Year, gas[country], label=country)

#print(gas.Year[::3])

plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figures.png', dpi=300)

plt.show()
```

## FIFA data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef') #google color picker

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill level')
plt.title('Distribution of Player Skills in FIFA 2018')


plt.show()
```

### Pie chart

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.1,0,0,.4)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.show()
```

### Box and Whisker

```python
fifa.head()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
realmadrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

plt.boxplot([barcelona, realmadrid, revs], labels = labels)

boxes = plt.boxplot([barcelona, realmadrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)

    # Set fill color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
