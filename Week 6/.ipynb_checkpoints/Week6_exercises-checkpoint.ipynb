```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Matplotlib

```python
ax= plt.figure()
ax.add_axes([0,0,1,1])

# data
x = np.arange(0,100)
y = x*2
z = x**2

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(x,y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_xticks([0,100])
axes2.set_yticks([0,200])



plt.show(axes1, axes2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,6))

ax[0].plot(x,y,'b--', lw=2)


ax[1].plot(x,z,'r', lw=2.5)
```

# Seaborn

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot(x='fare', y='age', data=titanic, kind='scatter')
```

```python
sns.distplot(titanic['fare'], kde=False, color='r', bins=30)
```

```python
sns.boxplot(x='class', y='age', data=titanic, hue='class', linewidth=2)
```
