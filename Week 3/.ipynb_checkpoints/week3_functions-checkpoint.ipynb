```python
# Turtle that draws square 

import turtle

# create turtle function to draw a square
def turtle_square(side_len):
    for i in range(4):
        turtle.forward(side_len)
        turtle.left(90)

# prompt the user to input how many times the user wants to draw square

print("Enter number of squares:")
n = int(input())

# draw a fan of n squares
for count in range(n):
    turtle_square(100)
    turtle.left(360/n)
turtle.hideturtle()
turtle.done()
```

```python
import turtle
def main():
    print("Please enter number of sqaures:")
    n = int(input())
    turtle_fan_of_squares(n, 100)
    turtle.hideturtle()
    turtle.done()
def turtle_square(side_len):
    for i in range(4):
        turtle.forward(side_len)
        turtle.left(90)
def turtle_fan_of_squares(num_of_sqrs, side_len):
    for count in range(num_of_sqrs):
        turtle_square(side_len)
        turtle.left(360/num_of_sqrs)
main()
```

## Converting Fahrenheit to Celsius

```python
# main

def main():
    print("Please enter a temperature in Fahrenheit:")
    fah_temp = float(input())
    cel_temp = converter(fah_temp)
    print(f"{fah_temp} Fahrenheit is {cel_temp:.3f} Celsius.")

def converter(fah_temp):
    C = (fah_temp - 32)*(5 / 9)
    return C

main()
```

## Palindrome Checker

```python
def main():
    print("Please enter a sentence:")
    sen = input()
    sen_ready = sen_prep(sen)
    sen_reverse = pal_check(sen_ready)
    if sen_reverse:
        print("Your sentence is a palidrome.")
    else: 
        print("Your sentence is not a palidrome.")
    sen_verified = sen_verification(sen_ready)
    if sen_reverse:
        print("Your sentence pass the test.")
    else: 
        print("Your sentence fail the test.")

def sen_prep(sen):
    only_let = keep_only_letter(sen)
    return only_let.lower()

def keep_only_letter(sen):
    res_str = ''
    for char in sen:
        if(char.isalpha() == True):
            res_str = res_str + char
    return res_str


def pal_check(sen):
    rev_str = reverse_str(sen)
    if(sen == rev_str):
        return True
    else:
        return False

def reverse_str(sen):
    res_str = ''
    for char in sen:
        res_str = char + res_str
    return res_str

def sen_verification(sen):
    verified_str = reverse_str(sen)
    if (sen == verified_str):
        return True
    else:
        return False

main()

    
```

```python
# time module has a sleep method which will help slow down the execution of loop
import time 
i = -2
while i < 5:
    i = i + 1
    try:
        print(1.0 / i)
        # This will halt the code for 2 seconds
        time.sleep(2)
    except KeyboardInterrupt:
        # Lets raise the exception that we just caught!
        raise KeyboardInterrupt('Ctrl C pressed')
    except ZeroDivisionError:
        print("So, you're trying to divide by zero huh?")
```

```python
import requests

# URL of the CSV file
url = 'https://www.hankyung.com/article/2024072846121'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content to a file
    with open('sample_articles.csv', 'wb') as file:
        file.write(response.content)
    print("File downloaded successfully.")
else:
    print(f"Failed to download file. Status code: {response.status_code}")
    
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the web page
url = 'https://www.hankyung.com/article/2024072846121'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract text from the web page
    # This might need to be customized based on the website's structure
    # For demonstration, we'll extract all paragraphs
    paragraphs = soup.find_all('p')

    # Save text to a file
    with open('extracted_text.txt', 'w', encoding='utf-8') as file:
        for paragraph in paragraphs:
            file.write(paragraph.get_text() + '\n')

    print("Text has been extracted and saved to 'extracted_text.txt'.")
else:
    print(f"Failed to retrieve the web page. Status code: {response.status_code}")
```
